# Image inverter

Use `cv2` (opencv library) to read jpg images

Use `matplotlib.pyplot` to show these images

Use `numpy` to convert cv2 images to arrays

```python
import cv2
# ! pip install opencv-python
# import matplotlib.image as mpimg

from matplotlib import pyplot as plt
import numpy as np
```

The following methods is used to make the conversion

The `imread` method allows you to split an image into its RGB factors

```python
image = cv2.imread("foto2.jpg")

# implot = cv2.imshow('image', image)
# cv2.waitKey()

# Turn BGR to RGB
imageRGB = np.array(image)

imageRGB[:,:,0] = image[:,:,2]
imageRGB[:,:,1] = image[:,:,1]
imageRGB[:,:,2] = image[:,:,0]

plt.imshow(imageRGB)

image = imageRGB
del imageRGB
```

> ### Warning!
> If you don't use the `cv2.waitkey()` line after `imshow` function in cv2, this will cause problems

Now, we will split this RGB factors and show them on the screen.

The main point in this section is that each factor will be showed in an gray scale.
Every pixel needs a 3-value array to have its colors `[255 255 255]` will be a white color and `[0 0 0]` will be a black color.

In the first section `r` value will be an 2d array in which every value will be the red value. To represent a color we need the 3 values so the gray scale will copy the red value in the other 2 factors.


So if the red value in the pixel with the `[i][j]` position is `k`, the gray scale image created only for red will have the next array in that position `[k k k]`
This will make a gray-scale representation in which the brightness value will be the red intensity.

```python
# Split colors in each RGB factors
r = image[:,:,0]
g = image[:,:,1]
b = image[:,:,2]

plt.subplot(1,3,1)
plt.imshow(r, cmap='gray')
plt.subplot(1,3,2)
plt.imshow(g, cmap='gray')
plt.subplot(1,3,3)
plt.imshow(b, cmap='gray')
plt.show()
```

```python
for j in range(800):
    for i in range(800):
        image[j,i,0] = 255 - r[j,i]
        image[j,i,1] = 255 - g[j,i]
        image[j,i,2] = 255 - b[j,i]

# cv2.imshow('image', image)
# cv2.waitKey()

plt.imshow(image)
```
