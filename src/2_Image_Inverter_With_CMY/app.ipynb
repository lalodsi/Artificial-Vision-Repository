# Image inverter using CMY colors

In this practice I'll get an BGR image, turn in into RGB, then into CMY colors, show every color and then use them to get the image inversion.

```python
import cv2 as opencv
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Getting image
image = opencv.imread("foto.jpg")

# Turn image into RGB factors
imageRGB = np.array(image, dtype=int)
imageRGB[:,:,0] = image[:,:,2]
imageRGB[:,:,2] = image[:,:,0]

image = imageRGB
del imageRGB

plt.imshow(image)
```

```python
# Get CMY factors
Cyan = 255 - image[:,:,0]
Magenta = 255 - image[:,:,1]
Yellow = 255 - image[:,:,2]

plt.subplot(141)
plt.imshow(image)
plt.axis('off')
plt.title("Original")

plt.subplot(142)
plt.imshow(Cyan, cmap='gray')
plt.axis('off')
plt.title("Cyan")

plt.subplot(143)
plt.imshow(Magenta, cmap='gray')
plt.axis('off')
plt.title("Magenta")

plt.subplot(144)
plt.imshow(Yellow, cmap='gray')
plt.axis('off')
plt.title("Yellow")

plt.show()
```

```python
inverted_image = np.zeros(image.shape, dtype=int)

inverted_image[:,:,0] = Cyan
inverted_image[:,:,1] = Magenta
inverted_image[:,:,2] = Yellow

plt.imshow(inverted_image)
```
