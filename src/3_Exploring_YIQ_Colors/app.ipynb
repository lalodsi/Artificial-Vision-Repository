# 3 Exploring YIQ Colors

YIQ Colors can be a new color tone


```python
import cv2 as opencv
from matplotlib import pyplot as plt
import numpy as np
```

Now we'll use the next array to realize a conversion between YiQ colors to RGB colors
<br>
<br>
![Rostro](matrix.svg)
<br>


```python
yiq_matrix = np.array([
    [0.299, 0.587, 0.114],
    [0.5957, -0.2744, -0.32126],
    [0.211, -0.522, 0.311]
])
```

The before array represents the next;
- Y I don't know
- I is for in-phase
- Q is for quadrature

```python
image = np.array(opencv.imread("foto2.jpg"))

size = image.shape

yiq_image = np.zeros(size)

[rows, cols, bits] = size
for j in range(rows):
    for i in range(cols):
        yiq_image[j,i] = yiq_matrix.dot(image[j,i])

Y = yiq_image[:,:,0]
I = yiq_image[:,:,1]
Q = yiq_image[:,:,2]

plt.subplot(131)
plt.imshow(Y, cmap='gray')
plt.axis('off')

plt.subplot(132)
plt.imshow(I, cmap='gray')
plt.axis('off')

plt.subplot(133)
plt.imshow(Q, cmap='gray')
plt.axis('off')
```

```python
# Comparison

gray = opencv.cvtColor(image, opencv.COLOR_RGB2GRAY)

plt.subplot(121)
plt.imshow(gray, cmap='gray')
plt.axis('off')
plt.title("With Open CV functions")

plt.subplot(122)
plt.imshow(Y, cmap='gray')
plt.axis('off')
plt.title("Gray scale Y component in YIQ colors")
```
