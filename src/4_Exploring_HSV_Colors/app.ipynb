```python
import numpy as np
import cv2 as opencv
from matplotlib import pyplot as plt
```

# Exploring HSV Colors

In this practice we'll use the HSV Conversion to show these colors in gray scale

```python
image = opencv.imread("foto.jpg");

imagenRGB = np.array(image, dtype=int)
imagenRGB[:,:,0] = image[:,:,2]
imagenRGB[:,:,2] = image[:,:,0]
image = imagenRGB

del imagenRGB

plt.imshow(image)
plt.show()
```

```python
image = np.array(image, dtype=np.uint8)
imageHSV = opencv.cvtColor(image, opencv.COLOR_RGB2HSV)

Hue = imageHSV[:,:,0]

plt.imshow(Hue, cmap='gray')
```

```python
Saturation = imageHSV[:,:,1]

plt.imshow(Saturation, cmap='gray')
```

```python
Value = imageHSV[:,:,2]

plt.imshow(Value, cmap='gray')
```
